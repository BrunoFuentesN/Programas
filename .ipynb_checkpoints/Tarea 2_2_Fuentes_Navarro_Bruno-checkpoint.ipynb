##### TAREA - Integración montecarlo

1. Realizar las siguientes integrales. Poner los resultados en una tabla cuyas filas correspondan a la cantidad de términos utilizados en la aproximación (usar 10, 100, 1000, 10000 y 100000 términos) y cuyas columnas correspondan a las funciones.
    + $\int_{1}^{3} e^{x}\log(x)\text{d}x$.
    + $\int_1^3 \sin(x)\ln(x) \text{d}x$

```python
#Importamos las librerias a utilizar

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Creamos la tabla y anexamos los resultados de la integral

N = np.logspace(1,5,5)
tabla=pd.DataFrame(columns=['Función 1','Función 2'],index=N)
def f1(x):
    return np.exp(x)*np.log(x)
def f2(x):
    return np.sin(x)*np.log(x)

def montecarlo1(f,a,b,N):
    u=np.random.uniform(a,b,size=N)
    y=f(u)
    return np.mean(y)*(b-a)
for n in N:
    tabla.loc[n,'Función 1']=montecarlo1(f1,1,3,int(n))
    tabla.loc[n,'Función 2']=montecarlo1(f2,1,3,int(n))
tabla
```

2. Aproximar el área de la región descrita por

$$x^2+2y^2\leq-2x+8y-7$$

Poner los resultados en una tabla cuyas filas correspondan a la cantidad de términos utilizados en la aproximación (usar 10, 100, 1000, 10000 y 100000 términos) y en las columnas el valor de aproximación y el error relativo.

Definir unos límites adecuados $a_1$, $b_1$, $a_2$ y $b_2$, para la región $R=\left[a_1,b_1\right]\times\left[a_2,b_2\right]$

```python
#Importamos las librerias a utilizar

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Definimos Montecarlo 2

def montecarlo2(region,xmin,xmax,ymin,ymax,N):
    x=np.random.uniform(xmin,xmax,N)
    y=np.random.uniform(ymin,ymax,N)
    B=region(x,y)
    AR=(xmax-xmin)*(ymax-ymin)
    return np.mean(B)*AR
```

```python
# Definimos la función a analizar

def region(x,y):
    return x**2+2*y**2<-2*x+8*y-7

# Se tiene que deducir los limites de integración para los limites de 'esta'

# Se proponen limites en'x'
x=np.random.uniform(-2.5,0.45,100)  

# Se proponen limites en'y'
y=np.random.uniform(1,3,100) 

# Se crean coordenadas para llenar el espacio requerido
X,Y=np.meshgrid(x,y)  

# Se grafica el espacio requerido rellenando la region 
plt.scatter(X,Y,c=~region(X,Y),cmap='coolwarm')
plt.grid()
```

#### Se hallaron límites:
#### Para $x$ de $-2.5$ a $0.45$ 
#### Para $y$ de $1$ a $3$

```python
# Creamos la tabla para anexar los resultados

N=np.logspace(1,5,5)
tabla=pd.DataFrame(columns=['Aproximación','Error'],index=N)
```

```python
#Estbalecemos el area de los limites(Conocida)
AR=(0.45-(-2.5))*(3-(-1)) 

# Agregamos los resultados y vemos el error de cada aproximación en la tabla

for n in N:
    tabla.loc[n,'Aproximación']=montecarlo2(region,-2.5,0.45,1,3,int(n))
    th=tabla['Aproximación'][n]/AR
    tabla.loc[n,'Error']=np.sqrt((1-th)/th)/(np.sqrt(n))
tabla
```
